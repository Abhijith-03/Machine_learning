```python
import pandas as pd 
import numpy as np
```

```python
df = pd.read_csv("data\Housing.csv")
df
```

```python
df.head(15)
```

```python
X = df.iloc[:,1:]
y = df.iloc[:,0]
```

```python
X
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(X, y, test_size = 0.2 , random_state = 0)
```

```python
X_train
```

```python
cat_cols = ['furnishingstatus','mainroad','guestroom','basement','hotwaterheating','airconditioning','prefarea']
```

```python
oe = OrdinalEncoder(categories=[['unfurnished','semi-furnished','furnished'],
['no','yes'],
['no','yes'],
['no','yes'],
['no','yes'],
['no','yes'],
['no','yes'],
])
```

```python
oe.fit(X_train[cat_cols])
```

```python
X_train[cat_cols] = oe.transform(X_train[cat_cols])
X_test[cat_cols] = oe.transform(X_test[cat_cols])
```

```python
X_train.head(10)
```

```python
X_test.head(10)
```

```python
from sklearn.linear_model import LinearRegression 
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import mean_absolute_error , mean_squared_error , r2_score
```

```python
print("MAE:",mean_absolute_error(y_test,y_pred))
print("MSE:",mean_squared_error(y_test,y_pred))
print("R2 Score:",r2_score(y_test,y_pred))
```
